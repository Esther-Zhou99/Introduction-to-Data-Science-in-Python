```python
def proportion_of_education():
    dict={}
    import pandas as pd
    df= pd.read_csv("NISPUF17.csv",usecols = ['EDUC1'])
    lst=list(df.value_counts())
    total=int(df.count())
    dict['less than high school'] =float(lst[3]/total)
    dict['high school'] =float(lst[2]/total)
    dict['more than high school but not college'] = float(lst[1]/total)
    dict['college'] =float(lst[0]/total)
    return dict
    raise NotImplementedError()
proportion_of_education()
```

```python
def average_influenza_doses():
    import pandas as pd
    df= pd.read_csv("NISPUF17.csv", usecols= ['CBF_01','P_NUMFLU'])
    df=df.fillna(0)
    s1=0
    s2=0
    for i in range(len(df)):
        if df.CBF_01[i]== 1:
            s1=s1+int(df.P_NUMFLU[i])
    for j in range(len(df)):
        if df.CBF_01[j] ==2:
            s2=s2+int(df.P_NUMFLU[j])
    a=s1/int(df[df.CBF_01 == 1].shape[0])
    b=s2/int(df[df.CBF_01 == 2].shape[0])
    return (round(a,1),round(b,1))
    raise NotImplementedError()
average_influenza_doses()
```

```python
def chickenpox_by_sex():
    import pandas as pd
    df= pd.read_csv("NISPUF17.csv", usecols= ['SEX','HAD_CPOX','P_NUMVRC'])
    df=df.fillna(0)

    dict={}

    yes1=0
    no1=0
    for i in range(len(df)):
        if df.SEX[i]==1 and df.P_NUMVRC[i]>=1 :
            if df.HAD_CPOX[i] ==1:
                yes1+=1
            if df.HAD_CPOX[i] ==2:
                no1+=1

    yes2=0
    no2=0
    for i in range(len(df)):
        if df.SEX[i]==2 and df.P_NUMVRC[i]>=1 :
            if df.HAD_CPOX[i] ==1:
                yes2+=1
            if df.HAD_CPOX[i] ==2:
                no2+=1

    dict['male']=round(yes1/no1,4)
    dict['female']=round(yes2/no2,4)

    return dict
    raise NotImplementedError()
```
